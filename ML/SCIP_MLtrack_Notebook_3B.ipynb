## A quick reminder about labels

A label is a known category or outcome attached to a sample. For example, a patient might have a known tumor subtype or a known outcome recorded in the dataset.

In unsupervised learning, we do not use labels to create the groups. We look for patterns in the feature data itself.

In this notebook, we will first build clusters from the METABRIC feature data. Later, we will use one outside biological label to help interpret what the clusters may mean.

## A simple framework for thinking about cluster evaluation

| Type of evidence | What it helps us check | Example in this notebook |
|---|---|---|
| Internal evidence | Whether clusters look reasonably separated or balanced | Silhouette score, cluster sizes |
| Comparison across methods | Whether two clustering methods suggest a similar broad story | K-means vs hierarchical clustering |
| Outside information | Whether clusters line up with meaningful biological patterns | PAM50 subtype |

## Step 6: Use PAM50 subtype to interpret the broad structure

PAM50 subtype is a good outside label to start with because it is closer to the biology of the tumor itself than a later clinical outcome would be.

To compare clusters with PAM50 subtype, we will use `pd.crosstab(...)`. This function makes a **contingency table**, which is a table showing how often two categories occur together.

Here:
- the **rows** will be clusters
- the **columns** will be PAM50 subtypes
- each cell will show how many patients fall into that combination

When we use `pd.crosstab(...)` by itself, it gives us a **count table**.

Then we will use `pd.crosstab(..., normalize="index")` to make a **row-proportion table**. In that version, each row adds up to 1, so the values show the proportions within each cluster.

```python
kmeans_pam50_counts = pd.crosstab(
    metabric_eval_df["kmeans_cluster"],
    metabric_eval_df["pam50_+_claudin-low_subtype"]
)

print("K-means PAM50 count table:")
kmeans_pam50_counts
```

```python
# Now adding normalize="index" to add up the rows to 1 and
# give proportions for each cell instead of raw counts.
kmeans_pam50_props = pd.crosstab(
    metabric_eval_df["kmeans_cluster"],
    metabric_eval_df["pam50_+_claudin-low_subtype"],
    normalize="index"
)

print("K-means PAM50 row-proportion table:")
display(kmeans_pam50_props.round(3))
```

In the row-proportion table, each row represents one cluster from K-means, and the values show the proportion of patients in that cluster who fall into each PAM50 subtype.

To interpret it, look across one row at a time and identify the largest values. Here, cluster 0 leans toward LumA, cluster 1 leans more toward Basal and Her2, and cluster 2 is split mostly between LumA and LumB.

Notice that the PAM50 variable has more categories than our 3-cluster solutions. That is not a problem. In unsupervised learning, the number of clusters does not need to match the number of outside labels. We are asking whether the clustering captures **any broad biological structure**, not whether it perfectly reproduces the PAM50 labeling system.
