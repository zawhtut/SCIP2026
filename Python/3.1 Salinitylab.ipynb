```python
# Step 1: Import the tools we need
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Step 2: Create a tiny sample dataset
practice_data = pd.DataFrame({
    "Date": ["2021-10-01", "2021-10-02", "2021-10-03"],
    "Salinity": [28.4, 29.1, 27.8],
    "Location": ["Fort Point", "Fort Point", "Fort Point"]
})

# Step 3: Look at the sample data
print("Practice dataset:")
display(practice_data)

# Step 4: Convert the Date column to datetime format
practice_data["Date"] = pd.to_datetime(practice_data["Date"])

print("\nPractice dataset after converting Date:")
display(practice_data)

# Step 5: Make a simple line plot

plt.figure(figsize=(7, 4))
plt.plot(practice_data["Date"], practice_data["Salinity"], marker="o")
plt.title("Practice Example: Salinity Over Time")
plt.axvline(pd.to_datetime("2021-10-02")) #adding a vertical line
plt.xlabel("Date")
plt.ylabel("Salinity (ppt)")
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d')) # Fix date formatting
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step 6: Interpret the result
print("Question: Which day had the highest salinity?")
print("Question: Did salinity go up or down overall?")
```
