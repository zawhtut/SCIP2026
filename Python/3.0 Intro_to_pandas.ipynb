# Intro to Pandas with Demo Data

## Purpose
This notebook teaches the basic commands you will use later in the salinity lab.

You will practice how to:
- create a simple dataset
- view the first and last rows
- check the size of the dataset
- inspect columns and data types
- calculate summary statistics
- make a simple graph

The data in this notebook is **demo data** only. It is not the real salinity dataset.

## Step 1: Import pandas

Pandas is a Python library used to work with tables and CSV files.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Step 2: Create demo data

We will make a small sample dataset so you can practice before using the real lab data.

```python
demo_data = pd.DataFrame({
    "Date": ["2021-10-01", "2021-10-02", "2021-10-03", "2021-10-04", "2021-10-05", "2021-10-06", "2021-10-07"],
    "Salinity": [28.4, 29.1, 27.8, 30.0, 29.5, 28.0, 24.8],
    "Location": ["Fort Point", "Fort Point", "Fort Point", "Fort Point", "Fort Point", "Fort Point", "Fort Point"]
})

demo_data
```

## Step 3: View the first rows with `.head()`

`.head()` shows the first 5 rows of a dataset. This helps you preview the data quickly.

```python
demo_data.head()
```

## Step 4: View the last rows with `.tail()`

`.tail()` shows the last 5 rows of a dataset.

```python
demo_data.tail()
```

## Step 5: Check the size with `.shape`

`.shape` tells you how many rows and columns are in the dataset.

```python
demo_data.shape
```

## Step 6: See the column names

Column names tell you what information is stored in the table.

```python
demo_data.columns
```

## Step 7: Check data types with `.info()`

`.info()` shows the data type of each column and whether any values are missing.

```python
demo_data.info()
```

## Step 8: Summarize the numbers with `.describe()`

This gives summary statistics such as the mean, minimum, and maximum.

```python
demo_data.describe()
```

## Step 9: Convert the Date column to datetime

When we work with dates in Python, it is helpful to convert them into datetime format.

```python
demo_data["Date"] = pd.to_datetime(demo_data["Date"])
demo_data
```

## Step 10: Make a simple line plot

Graphs help us see patterns in the data.

```python
plt.figure(figsize=(7, 4))
plt.plot(demo_data["Date"], demo_data["Salinity"], marker="o")
plt.title("Dummy Salinity Data")
plt.xlabel("Date")
plt.ylabel("Salinity (ppt)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
