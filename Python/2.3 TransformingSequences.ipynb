# Mutability

In Python, **mutability** describes whether an object can be changed after it is created.

- **Mutable objects** can be modified in place.
- **Immutable objects** cannot be changed after creation; a new object must be created instead.

This matters because it affects how variables behave when you update values, pass them into functions, or assign one variable to another.

Common examples:

- **Mutable:** lists, dictionaries, sets
- **Immutable:** strings, integers, tuples


#  Dictionaries

Dictionaries are **mutable**, just like lists. That means you can change a value inside a dictionary after it is created.

In this lesson, we will mostly focus on **strings** and **lists**. You do **not** need to learn dictionaries yet, but it is helpful to know that they can also change.

## Example

```python
favorite_food = {"Ava": "pizza", "Ben": "tacos"}

print("Before:", favorite_food)

favorite_food["Ava"] = "pasta"

print("After: ", favorite_food)
```

In the example above, only Ava's value changes. The dictionary itself stays the same object, but one of its values is updated.


````python
favorite_food = {"Ava": "pizza", "Ben": "tacos"}

print("Before:", favorite_food)

favorite_food["Ava"] = "pasta"

print("After: ", favorite_food)
````


A **dictionary** stores information as **key-value pairs**.


Why use a dictionary?

# Strings are immutable

When you work with strings, Python does not change the original string.

Instead, string operations create a **new** string value.

## Example

````python
word = "python"
print("Original:", word)

upper_word = word.upper()
print("Uppercase:", upper_word)
print("Still original:", word)

new_word = word + "!"
print("Concatenated:", new_word)
print("Still original again:", word)
````

Notice that `word` never changes.

That is why code like `word.upper()` is not enough by itself if you want to save the result. You must assign the new string to a variable.


````python
my_list = [1, 2, 3]
my_list.append(4)    # Modifying the list by adding an element
print(my_list)       # Output: [1, 2, 3, 4]
````

````python
my_string = "Hello"
new_string = my_string + " World"    # Creating a new string with the concatenated value
print(new_string)                    # Output: "Hello World"
````

The immutability of an object affects how Python handles assignments, copying, and passing objects as arguments to functions. Immutable objects (strings and tuples) are generally safer to use in scenarios where you want to ensure that the object's value remains unchanged, while mutable objects provide flexibility for modifying and updating data.

Understanding the mutability of objects is crucial when it comes to avoiding unexpected side effects, correctly passing and sharing data between functions, and managing memory efficiently in your Python programs.

# String indexing

You can access a single character in a string using the index operator `[]`.

Python counts from `0`, so the first character is at index `0`.

````python
text = "transform"
print(text[0])     # first character
print(text[3])     # fourth character
print(text[-1])    # last character
print(text[-3])    # third from the end
````

# String slicing

Slicing lets you extract part of a string.

The general form is:

```python
string[start:stop:step]
```

- `start` is included
- `stop` is excluded
- `step` is optional

````python
phrase = "transforming sequences"

print(phrase[0:10])     # characters from index 0 through 9
print(phrase[:10])      # same as above
print(phrase[11:])      # from index 11 to the end
print(phrase[::2])      # every other character
print(phrase[::-1])     # reverse the string
````

````python
fruit = ["banana", "apple", "cherry"]
print(fruit)

fruit[0] = "pear"
fruit[-1] = "orange"
print(fruit)
````

````python
alist = ['a', 'b', 'c', 'd', 'e', 'f']
alist[1:3] = ['x', 'y']
print(alist)
````

````python
alist = ['a', 'b', 'c', 'd', 'e', 'f']
alist[1:3] = []
print(alist)
````

# Searching in strings

Python gives you several ways to check whether text contains a value.

## Membership testing
Use `in` and `not in` when you only need a True/False answer.

````python
sentence = "Python makes string handling easy"

print("Python" in sentence)
print("easy" in sentence)
print("hard" not in sentence)
print("java" in sentence)
````

## Finding positions

Use `find()` or `index()` when you want the position of a substring.

- `find()` returns `-1` if the substring is not found
- `index()` raises an error if the substring is not found


````python
sentence = "Python makes string handling easy"

print(sentence.find("string"))
print(sentence.find("python"))   # not found because of case
print(sentence.index("easy"))

# Uncomment the next line to see the error that index() raises when the text is missing.
# print(sentence.index("python"))
````

# String formatting

The `format()` method lets you build a string from values.

This is useful when combining text and variables into a sentence.

````python
name = "Ava"
score = 93
grade = "A"

message = "Student: {}, Score: {}, Grade: {}".format(name, score, grade)
print(message)

# You can also reuse placeholders by position
sentence = "{0} scored {1} points. {0} was proud of that result.".format(name, score)
print(sentence)
````

# Strings versus lists

A list can be changed after it is created.

This means you can replace, add, or remove items without creating a brand-new list.

````python
fruits = ["banana", "apple", "cherry"]
print("Original list:", fruits)

fruits[0] = "pear"
fruits[-1] = "orange"

print("Updated list:", fruits)
````

# Replacing slices in a list

Lists support slice assignment.

This means you can replace several items at once.


````python
letters = ['a', 'b', 'c', 'd', 'e', 'f']
print("Before:", letters)

letters[1:3] = ['x', 'y']
print("After replacing a slice:", letters)

letters[1:3] = []
print("After deleting that slice:", letters)
````

# Deleting list elements

The `del` statement removes items from a list by position.

It is often clearer than assigning an empty list to a slice.

````python
items = ['one', 'two', 'three', 'four', 'five']
print("Original:", items)

del items[1]
print("After deleting one item:", items)

del items[1:3]
print("After deleting a slice:", items)
````

# The `append()` method

`append()` adds one item to the end of a list.

Important: `append()` changes the list in place and returns `None`.

````python
numbers = [1, 2, 3]
result = numbers.append(4)

print("List after append:", numbers)
print("Return value:", result)
````

## Common pitfall

A very common mistake is writing:

```python
numbers = numbers.append(4)
```

That makes `numbers` become `None`, because `append()` does not return the updated list.

Correct usage:

```python
numbers.append(4)
```

# `split()` and `join()`

These two methods are often used together.

- `split()` turns one string into a list of pieces
- `join()` turns a list of strings into one string


````python
data = "red,green,blue"
parts = data.split(",")

print("Split result:", parts)

rejoined = " | ".join(parts)
print("Joined result:", rejoined)
````

## More examples with `split()`

If you do not provide a separator, `split()` separates on whitespace.

````python
line = "   Python   is   fun   "
print(line.split())

csv_line = "math science history"
print(csv_line.split())
````

## More examples with `join()`

`join()` only works when every item in the list is already a string.

If the list contains numbers, you must convert them first.

````python
words = ["transform", "sequences", "with", "Python"]
print(" ".join(words))
print("-".join(words))

numbers = [1, 2, 3]
# print(",".join(numbers))  # This would raise a TypeError

number_text = [str(n) for n in numbers]
print(",".join(number_text))
````

# Aliasing

When two variables refer to the same list, they are aliases.

Changing one variable changes the shared object.

Since variables refer to objects, if we assign one variable to another, both variables refer to the same object.
image.png

````python
a = [81, 82, 83]
b = a

print("a is b:", a is b)

b[0] = 5

print("a:", a)
print("b:", b)
````

# Cloning lists

If you want a separate copy of a list, use cloning instead of assignment.

A common way to clone a list is with slicing:

```python
copy_list = original[:]
```

````python
original = [81, 82, 83]
clone = original[:]

print("Equal values:", original == clone)
print("Same object:", original is clone)

clone[0] = 5

print("Original after clone changes:", original)
print("Clone after changes:", clone)
````

# Comparing `is` and `==`

- `==` checks whether two values are equal
- `is` checks whether two variables refer to the same object

This distinction is very important when working with mutable objects like lists.

````python
x = [1, 2, 3]
y = [1, 2, 3]
z = x

print("x == y:", x == y)
print("x is y:", x is y)
print("x is z:", x is z)
````
