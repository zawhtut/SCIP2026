#Boolean Expressions
###A boolean expression is an expression that evaluates to a boolean (True or False) value. The equality operator, ==, compares two values and produces a boolean value related to whether the two values are equal to one another.
There are only two boolean values. They are True and False. Capitalization is important, since true and false are not boolean values (remember Python is case sensitive).
###Boolean values are not strings!

The == operator is one of six common comparison operators; the others are:

* x != y               # x is not equal to y
* x > y                # x is greater than y
* x < y                # x is less than y
* x >= y               # x is greater than or equal to y
* x <= y               # x is less than or equal to y

````python
#Predict which statement will return True and which will return False...
print(5 == 5) # 5 is equal to 5, True
print(5 == 6) #5 is equal to 6, False
print(5 >= 10)
print(5 <= 2)
print(5 != 7) #5 is not equal to 7, True

'''
write predicted output here

'''
````

# Logical operators
### There are three logical operators: and, or, and not. All three operators take boolean operands and produce boolean values. The semantics (meaning) of these operators is similar to their meaning in English:

* x and y is True if both x and y are True. Otherwise, and produces False.

* x or y yields True if either x or y is True. Only if both operands are False does or yield False.

* not x yields False if x is True, and vice versa.

````python
#The expression x > 0 and x < 10 is True only if x is greater than 0 and at the
#same time, x is less than 10. In other words, this expression is True if x is between 0 and 10, not including the endpoints.
x = 5
print(x > 0 and x < 10)

n = 25
print(n % 2 == 0 or n % 3 == 0)
````

#Precedence of Operators
##*Python will always evaluate the arithmetic operators first (** is highest, then multiplication/division, then addition/subtraction)*
The precedence of operators determines the order in which different operators are evaluated in an expression. It defines which operations are performed first and which are performed later.
It's important to note that when multiple operators have the same precedence, they are evaluated from left to right, unless overridden by parentheses. Additionally, understanding operator precedence helps in writing clear and unambiguous expressions and avoids unexpected results.

#Conditional Execution: Binary Selection

##*Selection statements, sometimes also referred to as conditional statements, give us this ability. The simplest form of selection is the if statement. This is sometimes referred to as binary selection since there are two possible paths of execution.*

```
# if BOOLEAN EXPRESSION:
    STATEMENTS_1        # executed if condition evaluates to True
else:
    STATEMENTS_2        # executed if condition evaluates to False
```
image.png


````python
import random #random module for computer to pick number in specified range

#randint() - (random interger range betweeen start and stop.)
#randint() -> (1-10) = |1,2,3...,10|
#range[]-> (1-10) = 1| 2,3,4...,9|10

#Generating/ build our actual guess model

random_number = random.randint(1,10) # the generate a number between 1 and 10 including 1 and 10.

while True: #while we a re guessing/ asking user's for thei guess
  guess = int(input("Guess the number (between 1 and 10): ")) #converting string input to be read by the computer as an integer!
  if guess == random_number:
    print("Congratulation! you've guessed the correct answer!")
    break #stop repeating the while statement.
  elif guess < random_number:
    print( "Too low! try again.")
  else:
    print("Too high! try again.")
````

#Omitting the else Clause: Unary Selection

##*Another form of the if statement is one in which the else clause is omitted entirely. This creates what is sometimes called unary selection. In this case, when the condition evaluates to True, the statements are executed. Otherwise the flow of execution continues to the statement after the body of the if.*

image.png

````python
temperature = 85

if temperature > 80:
  print("It's a hot day!")

print("stay Hydrated!")
````

The If statement checks if temperature > 80.
Since the condition is True, Python prints:
"It's a hot day!"


There is no else statement, so the condition were False, python would simply skip the if block and continue running the next line of code.

This is Unary Selection because there is only one possible action when the condition is true

#The Accumulator Pattern with Conditionals

The Accumulator Pattern is a programming technique where a variable keeps track of a running total, count, or result while a program loops through data.

When we combine this with conditionals (if Statements), we can choose which values should be added or counted.


Why is this useful?
In real programs, we often don't want every value, we only want the ones that match certain conditions.

For example:


*   Adding only positive numbers
*   Calculating the total cost of items over a certain price


Think of it like a filter: The loop looks at every item, the conditional decided if it should be included, and the accumulator stores the results



````python
s = "what if we went to the zoo"
x = 0
for i in s:
    if i in ['a', 'e', 'i', 'o', 'u']:
        x += 1
print(x)
````

#Accumulating the Max Value
##*find the maximum or minimum value. Instead of continuing to build up the accumulator value like we have when counting or finding a sum, we can reassign the accumulator variable to a different value.*


````python
nums = [9, 3, 8, 11, 5, 29, 2]
best_num = 0
for n in nums:
    if n > best_num:
        best_num = n
print(best_num)
````

````python
numbers = [1, 2, 3, 4, 5, 6]
total = 0

for num in numbers:
    if num % 2 == 0:
        total = total + num

print(total)
````
