```python
wcs = covid_data['state'].isin(['California', 'Washington', 'Oregon'])
```

```python
west_coast = covid_data.loc[wcs]
```

```python
sns.lineplot(data=west_coast, x='date', y='cases_avg_per_100k', hue='state')
plt.title('Covid in the West Coast')
plt.show()
```
